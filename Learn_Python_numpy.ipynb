```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients of the quadratic equation: ax^2 + bx + c
a = 1
b = -3
c = 2

# Create an array of x values
x = np.linspace(-10000, 10000 ,5)

# print (x)

# Calculate the corresponding y values using the quadratic equation
y = a*x**6 + b*x + c

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the Quadratic Equation')

# Add grid lines
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_tangent_curve(f, x, a):
    """
    Calculates the tangent line to the curve of function f at point a.

    Args:
        f (function): The function for which the tangent is calculated.
        x (numpy.ndarray): Array of x-values for plotting.
        a (float): The x-value at which to calculate the tangent.

    Returns:
        numpy.ndarray: Array of y-values for the tangent line.
    """
    print (type(f))

    h = 1e-6  # Small value for derivative calculation
    f_prime = (f(a + h) - f(a)) / h  # Approximate derivative

    print (f_prime)

    print ( type(f_prime))
    tangent = f(a) + f_prime * (x - a)
    return tangent

# Example usage:
def my_function(x):
    return np.cos(x)

x = np.linspace(-3, 9, 40)
a = 1.57  # Point at which to find the tangent

tangent = get_tangent_curve(my_function, x, a)
# print (tangent)

# Plotting
plt.plot(x, my_function(x), label='f(x)')
plt.plot(x, tangent, '--r', label=f'Tangent at x = {a}')
plt.scatter(a, my_function(a), color='black', label=f'Point of tangency ({a}, {my_function(a)})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tangent to a Curve')
plt.legend()
plt.grid(True)
plt.show()
```
